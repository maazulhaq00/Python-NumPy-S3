```python
import numpy as np
```

```python
x = np.array(
    [
        [11,26,39],
        [18,44,23],
        [17,87,91]
    ]
)
```

```python
# slicing & indexing

print(x[0][1])
print(x[0, 1])

print(x[0][-1])
print(x[-2][-2])


print(x[0][1:3])
print(x[0][1:])

print(x[0, 1:])

print(x[:, 0])
print(x[1, :])
print(x[1])


# 44 23 87 91
print(x[1:,1:])

# 11 26 18 44
print(x[:2,:2])
```

```python
x_flatten = x.flatten()
print(x_flatten)
```

```python
x_reshaped = x_flatten.reshape(3,3)
x_reshaped
```
